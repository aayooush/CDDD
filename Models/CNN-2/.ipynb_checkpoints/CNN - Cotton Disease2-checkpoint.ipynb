```python
import os
import cv2
import numpy as np
from time import time
from PIL import Image
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras import layers
from tensorflow.keras.models import Model
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.callbacks import TensorBoard
from sklearn.model_selection import train_test_split
%matplotlib inline
```

```python
gs = 0
N_CLASSES = 4   
IMG_W = 200
IMG_H = 200
disease ={0:'Healthy', 
          1:'Rust', 
          2:'Wooly_aphids', 
          3:'Mosaic_Virus'}   

def normalize(df):    
    return (df - df.min()) / (df.max() - df.min())
def denormalize(df,norm_data):    
    return (norm_data * (df.max() - df.min())) + df.min()
```

```python
def training_process():
    X = []
    y = list()
    category = {'Healthy':0, 'Rust':1, 'Wooly_aphids':2, 'Mosaic_Virus':3}
    root = 'CNN_Data/Training'
    print('-------------------- Preparing Data --------------------')
    for path, subdirs, files in os.walk(root):
        for name in files:
            if name[0]!='.':
                img_path = os.path.join(path,name)
                correct_cat = category[img_path.split('/')[-2]]
                print('Preparing File :- ',img_path, '\tAssigned Class', correct_cat)
                img = Image.open(img_path)
                img = img.resize((IMG_H,IMG_W))
                img_pixels = list(img.getdata())
                X.append(img_pixels)
                y.append(correct_cat)
                
    X = np.array(X)
    y = np.array(y)
    return X, y
```

```python
X, Y = training_process()
```

```python
data = []
for i in X:
    data.append(normalize(i.reshape(IMG_W,IMG_H,3)))
X = np.array(data)
```

```python
print('Data :- {}\nLabels :- {}'.format(X.shape,Y.shape))
```

```python
Y = to_categorical(Y)
```

```python
def alexnet(in_shape=(200,200,3), n_classes=N_CLASSES, opt='sgd'):
    in_layer = layers.Input(in_shape)
    conv1 = layers.Conv2D(96, 11, strides=4, activation='relu')(in_layer)
    pool1 = layers.MaxPool2D(3, 2)(conv1)
    conv2 = layers.Conv2D(256, 5, strides=1, padding='same', activation='relu')(pool1)
    pool2 = layers.MaxPool2D(3, 2)(conv2)
    conv3 = layers.Conv2D(384, 3, strides=1, padding='same', activation='relu')(pool2)
    conv4 = layers.Conv2D(256, 3, strides=1, padding='same', activation='relu')(conv3)
    pool3 = layers.MaxPool2D(3, 2)(conv4)
    flattened = layers.Flatten()(pool3)
    dense1 = layers.Dense(4096, activation='relu')(flattened)
    drop1 = layers.Dropout(0.5)(dense1)
    dense2 = layers.Dense(4096, activation='relu')(drop1)
    drop2 = layers.Dropout(0.5)(dense2)
    preds = layers.Dense(n_classes, activation='softmax')(drop2)

    model = Model(in_layer, preds)
    model.compile(loss="categorical_crossentropy", optimizer=opt,
	              metrics=["accuracy"])
    return model


model = alexnet()


model.summary()

tensorboard = TensorBoard(log_dir="./logs/{}".format(time()), 
                          histogram_freq=0, 
                          write_graph=True, 
                          write_grads=False, 
                          write_images=False, 
                          embeddings_freq=0, 
                          embeddings_layer_names=None, 
                          embeddings_metadata=None, 
                          embeddings_data=None, 
                          update_freq='epoch')

history = model.fit(X, Y,
                    epochs=200, verbose = 1,
                    callbacks = [tensorboard])
```

```python
model.save('CNN-Model{}.h5'.format(time()))
```

## Validation

```python
imgpath = []
def validation_process():
    X = []
    y = list()
    category = {'Healthy':0, 'Rust':1, 'Wooly_aphids':2, 'Mosaic_Virus':3}
    root = 'CNN_Data/Validation'
    print('-------------------- Preparing Data --------------------')
    for path, subdirs, files in os.walk(root):
        for name in files:
            if name[0]!='.':
                img_path = os.path.join(path,name)
                imgpath.append(img_path)
                correct_cat = category[img_path.split('/')[-2]]
                print('Preparing File :- ',img_path)
                img = Image.open(img_path)
                img = img.resize((IMG_H,IMG_W))
                img_pixels = list(img.getdata())
                X.append(img_pixels)
                y.append(correct_cat)
    X = np.array(X)
    y = np.array(y)
    return X, y
```

```python
valX, valY = validation_process()
```

```python
val_data = []
for i in valX:
    val_data.append(normalize(i.reshape(IMG_W,IMG_H,3)))
val_data = np.array(val_data)
```

```python
preds=model.predict(val_data)
```

```python
fp = []
for i in preds:
    fp.append(list(i).index(max(i)))
```

```python
pred = np.asarray(fp)
```

```python
for i in range(17,27):
    j = val_data[i]
    print("Prediction :- ", disease[pred[i]],"\tReality:",disease[valY[i]])
    plt.imshow(j)
    plt.show()
```

```python
f = pred == valY
print('Total :- ',len(f))
T = len(f[f == True])
F = len(f[f == False])
print('Correct Predictions:- {}\nFalse Prediction :- {}'.format(T, F))
error = F/(T+F) * 100
print('Error :- ', error)
accuracy = T/(T+F) * 100
print('Accuracy :- ', accuracy)
```

## TESTING

```python
import os
import cv2
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.models import load_model
%matplotlib inline

IMG_W = 200
IMG_H = 200

disease ={0:'Healthy', 
          1:'Rust', 
          2:'Wooly_aphids', 
          3:'Mosaic_Virus'} 

def normalize(df):    
    return (df - df.min()) / (df.max() - df.min())

def test_process(img_path):
    X = []
    img = Image.open(img_path)
    img = img.resize((IMG_H,IMG_W))
    img_pixels = list(img.getdata())
    X.append(img_pixels)
    X = np.array(X)
    return X

def predict_single_image(img_path):
    img = test_process(img_path)

    input_image = []
    input_image.append(normalize(img.reshape(IMG_W,IMG_H,3)))
    input_image = np.array(input_image)

    model = load_model('CNN-Model-2.h5')
    model.compile(loss="categorical_crossentropy", optimizer='sgd', metrics=["accuracy"])

    preds=model.predict(input_image)

    fp = []
    for i in preds:
        fp.append(list(i).index(max(i)))

    print("Prediction :- ", disease[fp[0]])
    plt.imshow(input_image[0])
    plt.show()   
    
img_path = 'CNN_Data/Test/healthy10.jpg'
predict_single_image(img_path)
```
