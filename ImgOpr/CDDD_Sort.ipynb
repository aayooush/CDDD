```python
import os
import tkinter 
from tkinter import *
from tkinter import filedialog
from PIL import ImageTk, Image
from shutil import copyfile,copy
```

```python
tk=tkinter.Tk()
tk.title("Picture Sort")
tk.geometry('800x500+80+80')

filename = ""
imgList = []
image_now = ""
ind = 0

def browsefunc():
    global filename
    global imgList
    filename = filedialog.askdirectory()
    for file in os.listdir(filename):
        if file.endswith(".jpg"):
            imgList.append(os.path.join(filename, file))
    pathlabel.config(text="Directory Selected :- " +filename)
    
def exit():
    tk.destroy()

def startfunc():
    global imgList
    global ind
#     print(imgList)
    
    print(imgList[ind])
    image = Image.open(imgList[ind]) 
    image = image.resize((200, 200), Image.ANTIALIAS)
    img = ImageTk.PhotoImage(image)
    l=Label(tk)
    l.config(image=img)
    l.image = img
    l.place(x = 300,y = 100)
    
    if ind == (len(imgList)-1):
        exit()
    
    def B1func():
        global ind
        copy(imgList[ind], 'C:\\Users\\NitishTalekar\\Desktop\\Sorting Pic\\B1') 
        ind = ind + 1
        startfunc()
    
    def B2func():
        global ind
        copy(imgList[ind], 'C:\\Users\\NitishTalekar\\Desktop\\Sorting Pic\\B2')
        ind = ind + 1
        startfunc()
        
    def B3func():
        global ind
        copy(imgList[ind], 'C:\\Users\\NitishTalekar\\Desktop\\Sorting Pic\\B3') 
        ind = ind + 1
        startfunc()
    
    def B4func():
        global ind
        copy(imgList[ind], 'C:\\Users\\NitishTalekar\\Desktop\\Sorting Pic\\B4')
        ind = ind + 1
        startfunc()
        
    def B5func():
        global ind
        copy(imgList[ind], 'C:\\Users\\NitishTalekar\\Desktop\\Sorting Pic\\B5') 
        ind = ind + 1
        startfunc()
    
    def B6func():
        global ind
        copy(imgList[ind], 'C:\\Users\\NitishTalekar\\Desktop\\Sorting Pic\\B6')
        ind = ind + 1
        startfunc()
        
    def B7func():
        global ind
        copy(imgList[ind], 'C:\\Users\\NitishTalekar\\Desktop\\Sorting Pic\\B7') 
        ind = ind + 1
        startfunc()
    
    def B8func():
        global ind
        copy(imgList[ind], 'C:\\Users\\NitishTalekar\\Desktop\\Sorting Pic\\B8')
        ind = ind + 1
        startfunc()
    
    B1 = Button(tk, text="B1", command=B1func)
    B1.place(x = 200,y = 350)
    
    B2 = Button(tk, text="B2", command=B2func)
    B2.place(x = 300,y = 350)
    
    B3 = Button(tk, text="B3", command=B3func)
    B3.place(x = 400,y = 350)
    
    B4 = Button(tk, text="B4", command=B4func)
    B4.place(x = 500,y = 350)
    
    B5 = Button(tk, text="B5", command=B5func)
    B5.place(x = 200,y = 450)
    
    B6 = Button(tk, text="B6", command=B6func)
    B6.place(x = 300,y = 450)
    
    B7 = Button(tk, text="B7", command=B7func)
    B7.place(x = 400,y = 450)
    
    B8 = Button(tk, text="B8", command=B8func)
    B8.place(x = 500,y = 450)
    

browsebutton = Button(tk, text="Browse", command=browsefunc)
browsebutton.pack()

pathlabel = Label(tk)
pathlabel.pack()

startbutton = Button(tk, text="Start", command=startfunc)
startbutton.pack()


tk.mainloop()
```
