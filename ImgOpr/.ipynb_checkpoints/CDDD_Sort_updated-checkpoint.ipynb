```python
import os
import tkinter 
from tkinter import *
from tkinter import filedialog
from PIL import ImageTk, Image
from shutil import copyfile,copy
```

```python
tk=tkinter.Tk()
tk.title("Picture Sort")
tk.geometry('800x700+80+80')

filename = ""
imgList = []
imgNo = []
image_now = ""
ind = 0

def browsefunc():
    global filename
    global imgList
    filename = filedialog.askdirectory()
    for file in os.listdir(filename):
        if file.endswith(".jpg"):
            imgList.append(os.path.join(filename, file))
            imgNo.append(int(file[4:-5]))
    pathlabel.config(text="Directory Selected :- " +filename)
    imgList = [x for _,x in sorted(zip(imgNo,imgList))]

def exit():
    tk.destroy()

def startfunc():
    global imgList
    global ind
#     print(imgList)
    print(ind)
    print(imgList[ind])
    image = Image.open(imgList[ind]) 
    image = image.resize((400, 400), Image.ANTIALIAS)
    img = ImageTk.PhotoImage(image)
    l=Label(tk)
    l.config(image=img)
    l.image = img
    l.place(x = 200,y = 100)
    imNo=Label(tk)
    imNo.config(text="ImgNo " + str(ind+1))
    imNo.place(x = 10,y = 100)
    
    if ind == (len(imgList)-1):
        exit()
    
    def B1func():
        global ind
        copy(imgList[ind], 'C:\\Users\\DEEPANSHU SONPAROTE\\Desktop\\finalData\\bud\\') 
        ind = ind + 1
        l.destroy()
        startfunc()
    
    def B2func():
        global ind
        copy(imgList[ind], 'C:\\Users\\DEEPANSHU SONPAROTE\\Desktop\\finalData\\bud_black\\')
        ind = ind + 1
        l.destroy()
        startfunc()
        
    def B3func():
        global ind
        copy(imgList[ind], 'C:\\Users\\DEEPANSHU SONPAROTE\\Desktop\\finalData\\flower\\') 
        ind = ind + 1
        l.destroy()
        startfunc()
    
    def B4func():
        global ind
        copy(imgList[ind], 'C:\\Users\\DEEPANSHU SONPAROTE\\Desktop\\finalData\\healthy green\\')
        ind = ind + 1
        l.destroy()
        startfunc()
        
    def B5func():
        global ind
        copy(imgList[ind], 'C:\\Users\\DEEPANSHU SONPAROTE\\Desktop\\finalData\\dry_yellow\\') 
        ind = ind + 1
        l.destroy()
        startfunc()
    
    def B6func():
        global ind
        copy(imgList[ind], 'C:\\Users\\DEEPANSHU SONPAROTE\\Desktop\\finalData\\red_patches\\')
        ind = ind + 1
        startfunc()
        
    def B7func():
        global ind
        copy(imgList[ind], 'C:\\Users\\DEEPANSHU SONPAROTE\\Desktop\\finalData\\whiteBrown_dots\\') 
        ind = ind + 1
        l.destroy()
        startfunc()
    
    def B8func():
        global ind
        copy(imgList[ind], 'C:\\Users\\DEEPANSHU SONPAROTE\\Desktop\\finalData\\discard\\')
        ind = ind + 1
        l.destroy()
        startfunc()
        
    def B9func():
        global ind
        copy(imgList[ind], 'C:\\Users\\DEEPANSHU SONPAROTE\\Desktop\\finalData\\red_brown_mix\\')
        ind = ind + 1
        l.destroy()
        startfunc()

        
    def B10func():
        global ind
        copy(imgList[ind], 'C:\\Users\\DEEPANSHU SONPAROTE\\Desktop\\finalData\\less_brown\\')
        ind = ind + 1
        l.destroy()
        startfunc()
        
        
    B1 = Button(tk, text="Bud", command=B1func)
    B1.place(x = 200,y = 350)
    
    B2 = Button(tk, text="black_bud", command=B2func)
    B2.place(x = 300,y = 350)
    
    B3 = Button(tk, text="Flower", command=B3func)
    B3.place(x = 400,y = 350)
    
    B4 = Button(tk, text="Healthy Green", command=B4func)
    B4.place(x = 500,y = 350)
    
    B5 = Button(tk, text="Dry Yellow", command=B5func)
    B5.place(x = 200,y = 450)
    
    B6 = Button(tk, text="Red Patches", command=B6func)
    B6.place(x = 300,y = 450)
    
    B7 = Button(tk, text="W-B dots", command=B7func)
    B7.place(x = 400,y = 450)
    
    B8 = Button(tk, text="Discard", command=B8func)
    B8.place(x = 500,y = 450)
    
    B9 = Button(tk, text="Red+Brown", command=B9func)
    B9.place(x = 200,y = 550)
    
    B10 = Button(tk, text="Less Brown", command=B10func)
    B10.place(x = 300,y = 550)
    

browsebutton = Button(tk, text="Browse", command=browsefunc)
browsebutton.pack()

pathlabel = Label(tk)
pathlabel.pack()

startbutton = Button(tk, text="Start", command=startfunc)
startbutton.pack()


tk.mainloop()
```
